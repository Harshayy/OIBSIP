# Exploration / Data Cleaning





```python
# Lets work with Age column (Single Column)
df['Age'].agg(['min','max'])
```

```python
def age_seg(a):
  if a <=30:
    return "20's Generation"
  elif a>30 and a<=40:
    return "30's Generation"
  elif a>40 and a<=50:
    return "40's Generation"
  elif a>50 and a<=60:
    return '50\'s Generation'
  elif a>60 and a<=70:
    return '60\'s Generation'
  elif a>70 and a<=80:
    return "70's Generation"
  else:
    return "80+ Generation"


df['age_seg']= df['Age'].apply(age_seg)

# view new column
df['age_seg']
```

```python
# Converting column data into meaningful names in the orginial dataframe
df['marital_Divorced'].replace({1:'Divorced'}, inplace=True)
df['marital_Married'].replace({1:'Married'},inplace=True)
df['marital_Single'].replace({1:'Single'},inplace=True)
df['marital_Together'].replace({1:'Together'},inplace=True)
df['marital_Widow'].replace({1:'Window'},inplace=True)
```

```python
# replacing 1 to string names, which we mentioned in above code
df['marital_status'] = df[['marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow']].apply(lambda row: next((value for value in row if value!= 0),0), axis=1)
df['marital_status']
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['marital_Divorced', 'marital_Married',
       'marital_Single', 'marital_Together', 'marital_Widow'], axis=1, inplace=True)
```

```python
df.head()
```

```python
sns.barplot(x='marital_status',y='Income',data=df,ci=None, palette='muted')
```

```python
# Assign columns into one column
df['education_Basic'].replace({1:'Basic'},inplace=True)
df['education_Graduation'].replace({1:'Graducation'},inplace=True)
df['education_Master'].replace({1:'Master'},inplace=True)
df['education_PhD'].replace({1:'PhD'},inplace=True)
```

```python
# processing column with Education_status
df['Education_status'] = df[['education_Basic','education_Graduation','education_Master','education_PhD']].apply(lambda row: next((value for value in row if value !=0),0), axis=1)
df['Education_status']
```

```python
df.drop(columns=['education_Basic','education_Graduation','education_Master','education_PhD'], axis=1,inplace=True)
```

```python
df['Education_status'].replace({0:'2nCycle'}, inplace=True)
```

This graph ilustrates that the more education an individual have, the more likely income they have.

# Feature the Data

```python
data_training = df[['Income','Education_status','marital_status','Age','child']]
data_training.head()
```

```python
data_training['Income'] = df['Income'].astype('int64')
```

```python
# giving labels to the columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_training['marital_status'] = le.fit_transform(df['marital_status'])
data_training['Education_status'] = le.fit_transform(df['Education_status'])
```

```python
data_training[['marital_status','Education_status']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# scaling the data
scaler = StandardScaler()
data_training[['Income','Education_status','marital_status','Age','child']] = scaler.fit_transform(data_training[['Income','Education_status','marital_status','Age','child']])
```

```python
data_training.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state= 42)
kmeans.fit(data_training)
data_training['Cluster_Group'] = kmeans.labels_
```

```python
data_training
```

```python
import plotly.express as px
```

```python
px.scatter(data_training, x='Education_status', y='Income',color= 'Cluster_Group', title='Clustering')
```

```python
# cluster with hover data
px.scatter(data_training,
                 x='Age',
                 y='Income',
                 color='Cluster_Group',
                 hover_data=['marital_status', 'Education_status'],
                 title='Cluster Scatter Plot with Details')
```
